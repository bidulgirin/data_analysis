# 데이터 수집

```python
url = "https://search.danawa.com/dsearch.php?query=%EB%AC%B4%EC%84%A0%EC%B2%AD%EC%86%8C%EA%B8%B0"

driver = webdriver.Chrome()
driver.get(url)
driver.implicitly_wait(3)
soup = BeautifulSoup(driver.page_source,"lxml")
```

```python
# 각각의 상품
prod_item = soup.select("div.main_prodlist li.prod_item")
```

```python
# 상품명
prod_item[0].select_one("a.click_log_product_standard_title_").string.strip()
```

```python
# 스펙목록
prod_item[0].select_one("div.spec-box--full div.spec_list").text.strip()
```

```python
# 가격
prod_item[0].select_one("a.click_log_product_standard_price_ strong").text.replace(",","").strip()
```

```python
prod_data = {
    "상품명": [],
    "스펙목록": [],
    "가격": [],
}
driver = webdriver.Chrome()
driver.get(url)


for i in range(10):
    time.sleep(2)
    try:
        soup = BeautifulSoup(driver.page_source, "lxml")
        prod_items = soup.select("div.main_prodlist li.prod_item")
        
        for item in prod_items:
           
                prod_data["상품명"].append(item.select_one("a.click_log_product_standard_title_").string.strip())
                prod_data["스펙목록"].append(item.select_one("div.spec-box--full div.spec_list").text.strip())
                prod_data["가격"].append(item.select_one("a.click_log_product_standard_price_ strong").text.replace(",","").strip())
    except:
        pass # 에러나면 넘어가~

    if i != 9:
        driver.find_element(By.CSS_SELECTOR, "a.now_spage + a").click()
        
```
