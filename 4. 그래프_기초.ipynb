# 막대 그래프

- 집단별 차이를 표현할 때 주로 사용
- 크고 작음의 차이를 한 눈에 파악하기 위해
- 예) 수량의 많고 적음 비교, 변화된 양에 대한 일별 , 월별 연별 통계등
- 가독성 면에서 막대 그래프는 일반적으로 항목의 개수가 적으면 가로 막대가 보기 좋고, 항목이 많으면 세로막대가 보기 편함

```python
x = np.arange(3)
years = ["2017", "2018", "2019"]
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years) # 눈금
plt.show()
```

```python
type(values)
```

```python
plt.bar(years,values) # x 에는 year 가 values 는 bar ...
plt.show()
```

```python
plt.bar(x,values, width = 0.6, align = "edge", color = "springgreen", edgecolor = "gray", linewidth = 3, 
        tick_label = years, log = True)
plt.show()
```

### point
- width : 막대의 넓이
- align =



- log : 변화량이 별로 없을때 로그 스케일로 표기하면 변화량을 나타내기 좋다!
- ex) 주식그래프

### 그룹 세로 막대 그래프

- 2016년 ~ 2019년 데이터를 막대 그래프로 그리기

```python
data = data.loc["Gyeonggi":"Daegu",2016:2019]
data
# Gyeonggi
# Incheon
# Busan
# Daegu
```

```python
#plt.bar(datas.index, datas.columns, width = 0.2, label = datas.columns)
index = np.arange(4)
plt.figure(figsize = (15,4))


for year in range(2016, 2020):
    chartdata = data[year]
    plt.bar(index, chartdata, width = 0.2, label = year) 
    index = index + 0.2 # [0.2, 1.2, 2.2, 3.2]  왜 0.2 냐? 막대그래프 width 가 0.2 이다

plt.ylim(35, 55)
plt.xlabel("area")
plt.ylabel("micrometer")
plt.title("2016~2019 Find Dust Group Bar Graph")
plt.xlabel("area")


# 그래프의 중앙에 눈금을 넣고 싶다

# index 는 array([0.8, 1.8, 2.8, 3.8]) 
# 막대 그래프는 0.2 width 를 가지므로 계산하면 중앙값은 0.3 이 되어야함
plt.xticks(index - 0.5)

plt.show()
```
