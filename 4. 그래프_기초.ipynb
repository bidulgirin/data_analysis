# 산점도 

- 서로 다른 두 연속형 변수 사이의 관계를 나타내며 연속형 변수의 상관성을 확인할 때 산점도 그래프를 사용
    - 예) 나이와 소득에 대한 상호 관련성 파악
    - 산점도에 표시되는 각 점들은 자료의 관측값을 의미하고 , 각 점의 위치는 관측값이 가지는 x축, y축 변수의 값으로 결정됨

```python
N = 50
x = np.random.rand(N)
y= np.random.rand(N)

plt.scatter(x,y)
plt.show()

print("x :" , x)
print("y :" , y)
```

```python
# 평면에서 x, y, z 의 관계를 생각하면 복잡할수있다.
# 3차원을 인지 할 수있지만 눈으로 받아드리는것은 평면적인 정보, 2차원 이다
# 직관적으로 3차원으로 파악하기는 어렵다
# 산점도는 3차원 2차원으로 바꿀수있다
# 있어보이게 3차원으로 보여줄수있구나~
```

```python
N = 50
x = np.random.rand(N)
y= np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2 # 크기
colors = np.random.rand(N)

plt.scatter(x,
            y,
            s=area, # 영역랜덤
            c=colors, # 변수에따라 color 값이 달라짐
            alpha = 0.5 # 투명도에 따라 
           )
plt.show()

print("area :" , area)
```

```python
# 실제값으로 산점도를 연습해보자
```

### 건강검진 데이터 남성과 여성의 허리둘레 히스토그램

```python
health = pd.read_excel("./data/health_screenings_2020_1000ea.xlsx")
# 보기 힘드니까 전처리
health["gender"] = health["gender"].map(lambda x :"Male" if x == 1 else "Female")
health["drinking"] = health["drinking"].map(lambda x : "None-drinking" if x == 0 else "Drinking")
# 이런 값들을 논의해서 전처리 해주세용
health["smoking"] = health["smoking"].map(lambda x : "Smoking" if x == 3 else "Non-smoking")
```

```python
health.loc[:,["gender","drinking","smoking","waist"]].tail()
```

```python
male_data = health.loc[health["gender"] == "Male",
    ["gender", "weight", "waist", "drinking", "smoking"]
]

female_data = health.loc[health["gender"] == "Female",
    ["gender", "weight", "waist", "drinking", "smoking"]
]
```

```python
male_data
```

```python
plt.figure(figsize = (10, 6))

sns.histplot(
    male_data["waist"], 
    kde = True,
    label = "male",
    alpha = 0.5,
    bins = 7,
    color="r" # 컬러도 지정
) 

sns.histplot(
    female_data["waist"], 
    kde = True, 
    label = "female",
    alpha = 0.5,
    bins = 7,
) 

plt.xlabel("waist")
plt.ylabel("count")
plt.legend()

plt.grid()
plt.show()
```

## 관계 그래프

- tips 컬럼,변수들의 관계들의 시각화를 보여준다

```python
fig = sns.pairplot(tips)
plt.show()
```

- 하삼각행렬과 상삼각행렬이 대칭
- 컬럼수가 늘면 보기 힘들다. 주로 작은 컬럼일때 사용해야함

```python
# 대칭이라는 점을 이용해서 다른 그래프를 그릴수있다 
pair_grid = sns.PairGrid(tips) # 그리드를 생성
pair_grid = pair_grid.map_upper(sns.regplot) # 산점도 그래프
pair_grid = pair_grid.map_diag(sns.histplot, kde = True) # 히스토그램, 밀집도
pair_grid = pair_grid.map_lower(sns.kdeplot, fill = True) #이차원 밀집도
plt.show()
```

```python
# 노란색 map_upper
# 빨간색 map_diag
# 검은색 map_lower
```

```python
sns.pairplot(tips, hue = "sex")
plt.show()
```

```python
# 높이가 아닌 모양새를 봐야함!!!!

# 높이를 안보는이유 : 남자의 수가 더 많을 수도 있으니까

# 쌍봉 : 우리가 알지 못하는 특이사항이 있음 => 봉오리를 나눌수 있는 기준을 찾기
# 관찰자가 여러명일때 쌍봉이 나올수도 있음
```

## FacetGrid 로 그룹별 그래프 그리기 ( 고수의 냄새를  )

```python
facet = sns.FacetGrid(
    tips, 
    col = "time"
)

facet.map(
    sns.histplot, # total_bill 가지고 히스토그램을 그려주겠다...
    "total_bill",
    kde = True, 
)

plt.show()
```

```python
facet = sns.FacetGrid(
    tips,
    col = "day", # 컬럼을 나눔 ( day 기준으로 )
    hue = "sex"
)
facet = facet.map(
    plt.scatter, # total_bill 데이터로 산점도를 그리겠다
    "total_bill",
    "tip"
)
facet = facet.add_legend() # 범례추가

plt.show()
```

```python
facet = sns.FacetGrid(
    tips,
    col = "time", # 컬럼을 나눔 ( day 기준으로 )
    row = "smoker",
    hue = "sex"
)
facet = facet.map(
    plt.scatter, # total_bill과 tip 데이터로 산점도를 그리겠다
    "total_bill",
    "tip"
)
facet = facet.add_legend() # 범례추가

plt.show()
```
