```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 히트맵

- 2개의 범주형 변수를 각각 x축 , y 축에 놓고 데이터를 행렬 형태로 분류하여 시각화

```python
# 피벗테이블로 범주형 변수를 각각ㄱ 행, 열로 재구분하여 정리
titanic = sns.load_dataset("titanic")

table = titanic.pivot_table(
    index = "sex", 
    columns = "class", 
    aggfunc = "size", # aggfunc : 합계, 개수, 표준편차, 분산 등의 산식
    observed = True)
```

```python
table
```

```python
# 엄청 마인크래프트같다...
# 데이터 개수가 많을때 효과적인 그래프이다...
```

```python
# 히트맵 그리기
sns.heatmap(
    table, # 데이터프레임
    annot = True, # 데이터 값 표시 여부
    fmt = "d", # 정수형 포멧
    cmap = "YlGnBu", # 컬러맵(공식홈페이지참고)
    linewidth = .1,  # 구분선
    #cbar = False # 컬러바 표시 여부
)

plt.show()
```

## 범주형 데이터의 산점도

- 산점도를 그릴때 데이터가 범주형이면 문제가 생길수있다
- 데이터가 몰려버려서 보기 힘들다
- 두 변수가 연속형일때 쓰는게 기본이다

### but! 
- 범주형 변수에 들어 있는 각 범주별 데이터의 분포를 확인
    - stripplot()
        - 데이터 분산 미고려 ( 중복 표시 잇음)
    - swarmplot()
        - 데이터의 분산을 고려하여 데이터 포인트가 서로 중복되지 않도록 시각화
        - 데이터가 퍼져 있는 정도를 입체적으로 볼 수 있음

```python
sns.set_style("whitegrid")

# 그래프 객체 생성
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# stripplot  - 데이터 분산 미고려 : 밧줄처럼 생김
sns.stripplot(
    x = "class",
    y = "age",
    data = titanic,
    ax = ax1,
    alpha = 0.3
)
# swarmplot - 데이터 분산 고려 (중복없음 = 겹쳐짐없음)
sns.swarmplot(
    x = "class",
    y = "age",
    data = titanic,
    ax = ax2
)

ax1.set_title("Stirp Plot")
ax2.set_title("Swarm Plot")

plt.show() 
# 12.7% 의 데이터는 모두 표시를 못하는 특성때문에 에러가 난다
```

## 막대 그래프

- seaborn 의 막대 그래프는 특별합니다.
- 범주형데이터의 막대 그래프를 그릴수있어요~~~~!!!
- 남자들의 생존률의 평균, 여자들의 생존률의 평균을 구한다.
- <small style="color:gray">y 축의 범주형 데이터를 이산형으로 임의적으로 변경하는것같다...</small>
- 검은색 바는 신뢰구간임

```python
# 그래프 객체 생성
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# x축에 y 축에 변수 할당
sns.barplot(
    x = "sex", # 범주형
    y = "survived", # 범주형
    data = titanic, 
    ax = ax1
)

# hue 옵션 추가
sns.barplot(
    x = "sex", 
    y = "survived", 
    hue ="class", # 또 나눠봤어용
    data = titanic, 
    ax= ax2
)

# 누적출력
sns.barplot(
    x = "sex", 
    y = "survived", 
    hue ="class", 
    data = titanic, 
    ax= ax3, 
    dodge = False, # 닷지를 금한다!!!!!!!!!!
)

# 차트 제목
ax1.set_title("titanic survived")
ax2.set_title("titanic survived - sex/class")
ax3.set_title("titanic survived - sex/class(stacked)")

# 퍼센트로 표시하려면 -------------------------------------
# 컬럼을 따로 만들어서 y 축 데이터로 쓰면 좋겠네용( 제일 이지 )

plt.show()
```

```python
# 신뢰구간 : 
# hue : 필요한거 추가해서 구분 색깔로 구분됨
# dodge = False : 스택으로 다같은 팀이된 f1 ...을 생각해라...그는 신이야!
```

## 빈도 그래프

- 별도의 전처리 없이 바로 컬럼을 넣을수있다!
- hue 같은걸로 범주를 나눌수있다

```python
# 그래프 객체 생성
fig = plt.figure(figsize = (15, 5))
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# 기본값
sns.countplot(
    x = "class",
    palette = "Set1", # 팔레트 옵션을 넣으면 컬럼별 색상이
    data = titanic,
    ax = ax1
)

# hue 옵션에 "who"추가
sns.countplot(
    x = "class",
    hue = "who",
    palette = "Set2",
    data = titanic,
    ax = ax2
)
# dodge = False 추가
sns.countplot(
    x = "class",
     hue = "who",
    palette = "Set3",
    data = titanic,
    dodge = False,
    ax = ax3
)

# 차트 제목
ax1.set_title("titanic class")
ax2.set_title("titanic class - who")
ax3.set_title("titanic class - who(stacked)")

plt.show()
```

## 조인트 그래프

```python
# 조인트 그래프 - 산점도
j1 = sns.jointplot(
    x = "fare",
    y = "age",
    data = titanic
)

j1.fig.suptitle("titanic fare - scatter", size = 15)
plt.show()
```

```python
# 조인트 그래프 - 회귀선
j1 = sns.jointplot(
    x = "fare",
    y = "age",
    data = titanic,
    kind = "reg" # 회귀선
)

j1.fig.suptitle("titanic fare - reg", size = 15)
plt.show()
```

```python
# 조인트 그래프 - 육각형
j1 = sns.jointplot(
    x = "fare",
    y = "age",
    data = titanic,
    kind = "hex" # 육각그래프
)

j1.fig.suptitle("titanic fare - hex", size = 15)
plt.show()
```

```python
# 조인트 그래프 - 터널 밀집 그래프
j1 = sns.jointplot(
    x = "fare",
    y = "age",
    data = titanic,
    kind = "kde" # 터널 밀집 그래프
)

j1.fig.suptitle("titanic fare - kde", size = 15)
plt.show()
```
