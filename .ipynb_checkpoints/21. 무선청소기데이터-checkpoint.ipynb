## 사용시간 단위 통일

- 분 단위로 통일
    - "시간" 단어가 있으면
        - "시간" 앞에 숫자를 추출한 뒤 60을 곱해 사용시간에 더함
        - "시간" 뒤의 글자에서 "분" 굴자 앞의 숫자를 추출해서 사용시간에 더함
 
    - "시간" 단어가 없으면
        - "분" 글자 앞의 숫자를 추출해서 사용시간에 더함
 
    - 위의 방식으로 계산되지 않을 경우 예외

```python
def convert_time_minute(time):
    try:
        if "시간" in time:
            hour = time.split("시간")[0]

            if "분" in time:
                minute = time.split("시간")[-1].split("분")[0]

            else : 
                minute = 0

        else:
            hour = 0
            minute = time.split("분")[0]
            if "~" in minute:
                minute = minute.split("~")[1]


        return int(hour) * 60 + int(minute)

    except:
        return None       
```

```python
new_use_time_list = []

for i in use_time_list:
    new_use_time_list.append(convert_time_minute(i))  
df["사용시간"] = new_use_time_list
```

### 흡입력 단위 통일
- 사용하는 단위 AW, W, PA
- 대략 1w = 1aw = 100pa 으로 단위 환산

- 위 전환식으로 AW 기준으로 단위를 통일
- 전처리 과정
- 모든 알파벳 문자를 대문자로 수정
- 흡입력에 "AW" 나 "W" 가 있으면 삭제
    - "," 도 삭제한뒤 숫자형 데이터로 변환
- 흡입력에 "PA" 글자가 있다면
    - "," 도 삭제한뒤 숫자형 데이터로 변환 후 100으로 나눔
 
- 흡입력 값이 비어 있거나 단위 변환 시 에러가 날 경우 예외처리

```python
def get_suction(value):
    try:
        value = value.upper()
        
        if "AW" in value or "W" in value:
            result = value.replace("A", "").replace("W","")
            result = int(result.replace(",", ""))
            

        elif "PA" in value:
            result = value.replace("PA", "")
            result = int(result.replace(",","")) / 100

        else:
            result = None

        return result

    except:
        return None
```

```python
new_suction_list = []
for i in suction_list:
    new_suction_list.append(get_suction(i))  
    
new_suction_list[:5]
```
