## 데이터 속성 탐색

- 상관 분석
    - 두 변수 간의 선형적 관계를 상관 계수로 표현 하는 것
    - 상관 계수는 공분산의 개념을 포함함
        - 공분산
            - 2개의 확률 변수에 대한 상관 정도
            - 2개의 변수 중 하나의 값이 상승하는 경향을 보일 때 다른 값도 상승하는 경향을 수치로 나타낸 것
            - 공분산 만으로 상관 관계를 구한다면 두 변수의 단위 크기에 영향을 받음
            - 큰 단위의 데이터라면 공분산이 커짐
     
        - 따라서 상관 계수는 공분산을 -1 과 1사이의 값으로 변환한 것
     
- 상관 관계를 통계적으로 탐색하는 방법
    - 단순 상관 분석: 피처가 2개일 때 상관 계수를 계산
    - 다중 상관 분석 : 대상 피처가 여러 개일 때 상호 간의 연관성 분석

### 아프리카와 유럽 간의 맥주 소비량 차이 검정

```python
# AF EU
af = drinks[drinks["continent"] == "AF"]
eu = drinks[drinks["continent"] == "EU"]
```

```python
# 정규성 검정
# shapiro-wilk(H0 : 정규분포를 따른다)
# stats 는 import 해온 통계용 모듈임
stats.shapiro(af["beer_servings"])
```

```python
# p-value 귀무가설이 기각되니 정규분포가 아니다!
```

```python
stats.shapiro(eu["beer_servings"])
```

```python
#  pvalue 가 0.1 이므로 eu는 정규분포이다
```

```python
# 지금은 정규분포로 만드는것을 생략한다
```

```python
# 등분산 검정
# bartlett(H0 : 두 집단이 분산이 같다)
stats.bartlett(af["beer_servings"], eu["beer_servings"])
```

```python
# p-value = 0.0592, 아슬아슬하게 등분산성이 같다고 한다
```

```python
# t-test(두 집단의 분산이 같은 경우)
stats.ttest_ind(
                af["beer_servings"], eu["beer_servings"],  
                equal_var = True # equal_var 같은것이 귀무가설
               ) 
```

```python
# p-value 를 보니 분산이 같구나~ 귀무가설 채택 
# df 자유도 96.0 데이터개수 97개 
```

- t-statistics(t-test의 검정 통계량)
    - 그 자체로는 아무 의미 없으며 p-value와 함께 해석해야함
    - 귀무가설과의 차이를 뜻함
- p-value(유의확률)
    - 가설이 얼마나 믿을 만한 것인지 를 나타내는 지표
    - 데이터를 새로 샘플링 했을 때 귀무 가설이 맞다는 전제 하에 현재 나온 통계값 이상이 나올 확률
        - 즉 p-value 가 낮으면 귀무 가설이 일어날 확률이 너무 낮기 때문에 귀무가설을 기각하게됨
        - 보통은 그 기준은 0.05이나 0.01을 기준으로 함
- 귀무가설
    - 처음부터 버릴 것을 예상하는 가설
    - 가설이 맞지 않다는 거을 증명하기 위해 수립하는 가설
    - 반대되는 가설을 대립 가설이라고 부르며, 귀무 가설이 거짓인 경우에 대안으로 참이 되는 가설임

#### 분석 내용
- t-test 의 귀무가설 : "두 집단의 평균이 같다"

- p-value 가 0.01 이하로 나타나 귀무가설 기각

- 따라서 "아프리카와 유럽 대륙 간의 맥주 소비량의 차이" 는 통계적으로 유의함
    - 두 집단의 평균은 다르다
